```
def preproceso(img):
  imgYCC = cv2.cvtColor(img, cv2.COLOR_BGR2YCR_CB)
  Gausianas = cv2.GaussianBlur(cv2.cvtColor(imgYCC, cv2.COLOR_BGR2GRAY),(5,5),0)
  ret,umbralizada = cv2.threshold(Gausianas,128,255,cv2.THRESH_BINARY_INV)
  Enmascarada = cv2.bitwise_and(img,img, mask = umbralizada)
  return imgYCC,Gausianas,umbralizada,Enmascarada
  # Se comienza sacar carateristicas 

def imprimir(img):
  plt.figure(figsize=(5,5))
  plt.imshow(img)
  plt.show()

def imprimir_dos(img1,img2):
  plt.figure(figsize=(5,5))

  plt.subplot(1,2,1)
  plt.imshow(img1)
  
  plt.subplot(1,2,2)
  plt.imshow(img2)
  plt.show()

def imprimir_tres(img1,img2,img3):
  plt.figure(figsize=(12,5))
  
  plt.subplot(1,3,1)
  plt.imshow(img1)

  plt.subplot(1,3,2)
  plt.imshow(img2)

  plt.subplot(1,3,3)
  plt.imshow(img3)
  plt.show()


def imprimir_cuatro(img1,img2,img3,img4):
  plt.figure(figsize=(15,5))
  
  plt.subplot(1,4,1)
  plt.imshow(img1)

  plt.subplot(1,4,2)
  plt.imshow(img2)

  plt.subplot(1,4,3)
  plt.imshow(img3)
  
  plt.subplot(1,4,4)
  plt.imshow(img4)

  plt.show()
```

```
img = cv2.imread('E17.jpg') # Read temporal image
imgYCC,Gausiana,umbralizada,Enmascarada = preproceso(img)
imprimir_cuatro(img,imgYCC,umbralizada,Enmascarada)
```

# Reduccion dimensional 
---





```
import torch
subset = df[['MoB', 'MoG','MoR', 'MeB','MeG', 'MeR','sdtB',
           'stdG','stdR','MoB2', 'MoG2','MoR2', 'MeB2','MeG2',
           'MeR2','sdtB2', 'stdG2','stdR2','hu1','hu2','hu3',
           'hu4','hu5','hu6','hu7','Ncont','MinArea','MaxArea',
           'ContConv','MinPer','MaxPeri','type']]

plantas  = torch.tensor(subset.values).float()
data = plantas[:,:-1]
target = plantas[:,-1]
# Reduccion de dimension
scaler = StandardScaler()
X = scaler.fit_transform(plantas)
NumberComponents = 30 # Numero de componenete
pca = PCA(n_components=NumberComponents)
principal_components = pca.fit_transform(X)
new_data = pd.DataFrame(data = principal_components)

plt.figure(figsize=(25,5))
plt.subplot(1,2,1)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('numero de componentes',size ="17")
plt.ylabel('varianza acumulada',size ="17");
plt.grid(color='b', linestyle='-', linewidth=0.5)
plt.title('varianza con 30 componentes', fontdict=None, loc='center', size = '20')


scaler = StandardScaler()
X = scaler.fit_transform(plantas)
NumberComponents = 8 # Numero de componenete
pca = PCA(n_components=NumberComponents)
principal_components = pca.fit_transform(X)
new_data = pd.DataFrame(data = principal_components)
new_data.insert(8, 'type',target, True)
new_data.drop(29)

# Evaluacion cambio varianza con 12 
plt.subplot(1,2,2)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('numero de componentes',size ="17")
plt.ylabel('varianza acumulada',size ="17");
plt.grid(color='b', linestyle='-', linewidth=0.5)
plt.title('varianza con 8 componentes', fontdict=None, loc='center', size = '20')
plt.show()
```

```
np.random.seed(23)
X_train, X_test, y_train, y_test =train_test_split(new_data, target, 
                                            test_size = 0.30, random_state = 28)

def imprimirResultados(clf,X_test,y_test,X_train,y_train):
  algoritmo = clf
  algoritmo.fit(X_train,y_train)
  algoritmo.predict(X_test)
  preds = algoritmo.predict(X_test)
  
  plt.figure(figsize=(15,6))
  plt.subplot(211)
  plt.hist([preds,y_test])
  plt.grid(color='b', linestyle='-', linewidth=0.5)
  plt.show()
```

```
clf =  KNeighborsClassifier(n_neighbors=3)
print(" 3 vecinos")
imprimirResultados(clf,X_test,y_test,X_train,y_train)
preds1 = clf.predict(X_test)
# pd.crosstab(y_test, preds, rownames=['real'], colnames=[' predicha'])

print(" 4 vecinos")
clf =  KNeighborsClassifier(n_neighbors=4)
imprimirResultados(clf,X_test,y_test,X_train,y_train)
preds2 = clf.predict(X_test)
# pd.crosstab(y_test, preds, rownames=['real'], colnames=[' predicha'])

print(" 5 vecinos")
clf =  KNeighborsClassifier(n_neighbors=5)
imprimirResultados(clf,X_test,y_test,X_train,y_train)
preds = clf.predict(X_test)
```

# Red Neuronales
---


```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping

X_test = np.asarray(X_test)
y_test = np.asarray(y_test)
X_train = np.asarray(X_train)
y_train = np.asarray(y_train)
# Build the neural network
model = Sequential()
model.add(Dense(1, input_dim=new_data.shape[1], activation='relu')) # Hidden 1
model.add(Dense(20, activation='relu')) # Hidden 2
model.add(Dense(1)) # Output
model.compile(loss='mean_squared_error', optimizer='adam')
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, 
                        patience=5, verbose=1, mode='auto',
                        restore_best_weights=True)

model.fit(X_train,y_train,validation_data=(X_test,y_test),callbacks=[monitor],
                                            verbose=2,epochs=1000)

from sklearn import metrics

# Predict
pred = model.predict(X_test)

# Measure MSE error.  
score = metrics.mean_squared_error(pred,y_test)
print("Final score (MSE): {}".format(score))
```

```
def chart_regression(pred, y, sort=True):
    t = pd.DataFrame({'pred': pred, 'y': y.flatten()})
    if sort:
        t.sort_values(by=['y'], inplace=True)
    plt.plot(t['y'].tolist(), label='expected')
    plt.plot(t['pred'].tolist(), label='prediction')
    plt.ylabel('output')
    plt.legend()
    plt.grid(color='b', linestyle='-', linewidth=0.5)
    plt.show()
chart_regression(pred.flatten(),y_test)
```
